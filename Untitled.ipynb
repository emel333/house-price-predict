Inspecting the Distribution of Model Errors: Q-Q Plotting
--

Now, I'll do some QQ plot analysis:

```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as smf
import scipy.stats as stats
plt.style.use('ggplot')


f = 'price~scaled_sqft_living'
f2 = 'price~scaled_sqft_living15'
f3 = 'price~condition_3'
f4 = 'price~baths_4_25'
f5 = 'price~baths_plus_reno'
f6 = 'price~home_age'
f7 = 'price~has_basement'
f8 = 'price~has_waterfront'

model_01 = smf.ols(formula=f, data=train3).fit()
model_02 = smf.ols(formula=f2, data=train3).fit()
model_03 = smf.ols(formula=f3, data=train3).fit()
model_04 = smf.ols(formula=f4, data=train3).fit()
model_05 = smf.ols(formula=f5, data=train3).fit()
model_06 = smf.ols(formula=f6, data=train3).fit()
model_07 = smf.ols(formula=f7, data=train3).fit()
model_08 = smf.ols(formula=f8, data=train3).fit()

resid1 = model_01.resid
resid2 = model_02.resid
resid3 = model_03.resid
resid4 = model_04.resid
resid5 = model_05.resid
resid6 = model_06.resid
resid7 = model_07.resid
resid8 = model_08.resid

fig = sm.graphics.qqplot(resid1, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid2, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid3, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid4, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid5, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid6, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid7, dist=stats.norm, line='45', fit=True)
fig = sm.graphics.qqplot(resid8, dist=stats.norm, line='45', fit=True)
```
